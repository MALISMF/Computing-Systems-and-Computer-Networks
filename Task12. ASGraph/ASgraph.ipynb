```python
graph = nx.read_edgelist("as20000102.txt", nodetype=int)
```

```python
graph_degree = graph.degree()
sorted_graph_degree = sorted(graph_degree, key=lambda x: x[1], reverse=True)
sorted_graph_degree[:10]
```
