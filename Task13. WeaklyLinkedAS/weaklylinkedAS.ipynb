```python
import networkx as nx
G = nx.read_edgelist('./as20000102.txt')
```

```python
found = False
for src_node, paths_dict in nx.shortest_path_length(G):
  if found: break
  for dest_node, shortest_path in paths_dict.items():
    if found: break
    if shortest_path == 1:
      for path in nx.all_simple_paths(G, source=src_node, target=dest_node):
        if len(path) > 10:
          print(f'Найдены слабо связанные вершины {src_node} и {dest_node}, которые связаны между собой прямым путём и путём длиной в {len(path)-1} рёбер:\n{' -> '.join(path)}')
          found = True
          break
```
