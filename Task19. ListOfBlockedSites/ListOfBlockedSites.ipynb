```python
import json
import os

file_path = 'domains_data.json'

if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' was not found. Please ensure the previous cell has been executed successfully.")
else:
    with open(file_path, 'r', encoding='utf-8') as f:
        domains_list = json.load(f)

    https_domains = [domain for domain in domains_list if 'https' in domain.lower()]

    print(f"Total domains loaded: {len(domains_list)}")
    print(f"Domains containing 'https': {len(https_domains)}")
    print("First 5 domains containing 'https':")
    for domain in https_domains[:5]:
        print(domain)
```

```python
import whois
import time
import whois.exceptions

available_domains = []

print(f"Starting WHOIS queries for {len(https_domains)} HTTPS domains. This may take a while...")

for i, domain_name in enumerate(https_domains[:50]):
    cleaned_domain = domain_name.strip('"').strip()

    if not cleaned_domain:
        continue

    try:
        domain_info = whois.whois(cleaned_domain)

        if domain_info and domain_info.expiration_date is None and \
           domain_info.registrar is None and \
           domain_info.creation_date is None:
            available_domains.append(cleaned_domain)
            print(f"[{i+1}/{len(https_domains[:50])}] Found available: {cleaned_domain}")
        else:

            pass
    except whois.exceptions.WhoisDomainNotFoundError:
        available_domains.append(cleaned_domain)
        print(f"[{i+1}/{len(https_domains[:50])}] Found available (no WHOIS entry): {cleaned_domain}")
    except Exception as e:
        print(f"[{i+1}/{len(https_domains[:50])}] Error querying {cleaned_domain}: {e}")
        pass

    time.sleep(0.5)

print(f"\nFinished WHOIS queries.")
print(f"Total HTTPS domains found available for registration: {len(available_domains)}")

if available_domains:
    print("Available domains:")
    for domain in available_domains:
        print(domain)
else:
    print("No HTTPS domains found available for registration among the processed domains.")
```
